```python
import pandas as pd # подключим библиотеку
import numpy as np

#Загружаем исходные данные
data_churn=pd.read_csv('data/Churn_Modelling.csv')
data_churn
```

```python
#Задание 1: Тщательно изучите документацию на библиотеку pandas и попробуйте самостоятельно выполнить группировку по другим столбцам.

#Проведём группировку по возрастам и выведем характеристики Баланса для них
ages = data_churn.sort_values(by=['Age'])\
    .groupby(by=['Age'],as_index=False)\
    ['Balance'].agg(['min','max','mean'])
print(ages)
#with pd.option_context('display.max_rows', None, 'display.max_columns', None):  # more options can be specified also
#    print(ages)
```

```python
#Задание 2: В качестве упражнения выведите на экран название (Name) фирмы, у которой цена (Price) входит в интервал Exclusive
#В наборе данных из примера нет фирм. Поэтому выполним данное действие для выборки Churn_Modelling.csv.
#Сначала приведём столбец Balance к текстовому виду и поделим всех клиентов на пять категорий: Low, Medium, High, Very High, Exclusive
cuted=pd.cut(data_churn['Balance'], 5,labels=['Low','Medium','High','Very High','Exclusive'], include_lowest=True)
data_churn_c = data_churn
data_churn_c['Balance']=cuted

#Теперь находим человека с Балансом эксклюзивного уровня
ExclusiveCustomer = data_churn[(data_churn['Balance'] == "Exclusive")].iloc[0]
print(ExclusiveCustomer.Surname)
```
