```python
# подключим библиотеки
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

#Загружаем выборку wine.csv и разбиваем её на подмножества X,y
data_wine=pd.read_csv('data/wine.csv')
X=data_wine.iloc[:,0:11]
y=data_wine['quality']
X,y
```

```python
#Проверим баланс выборки
target_count = data_wine.quality.value_counts()
print(target_count)
print('Good:', target_count[0])
print('Bad:', target_count[1])
print('Пропорция:', round(target_count[0] / target_count[1], 2), ': 1')

sns.countplot(x='quality', data= data_wine, palette="OrRd")
plt.box(False)
plt.xlabel('quality',fontsize=11)
plt.ylabel('Количество наблюдений',fontsize=11)
plt.show()
#Выборка уже является более менее сбалансированной. Хороших вин в выборке немного больше, чем плохих.
```

```python
#Прооводим андерсэмплинг методом Tomek Links.
from imblearn.under_sampling import TomekLinks

#Используем метод Tomek Links
tl = TomekLinks(sampling_strategy='majority')
X_res, y_res = tl.fit_resample(X, y)
print('Размер исходной выборки', X.shape)
print('Размер выборки после андерсемплинга',X_res.shape)
data_new=pd.concat([X_res, y_res], axis=1)
print(data_new.shape)

#Проверим баланс выборки
sns.countplot(x='quality', data= data_new, palette="OrRd")
plt.box(False)
plt.xlabel('quality',fontsize=11)
plt.ylabel('Количество наблюдений',fontsize=11)
plt.show()
#Метод Tomek Links привёл выборку к более сбалансированному состоянию, чем было изначально. 
```

```python
#Попробуем другой метод андерсемплинга
# перемешаем датафрейм
shuffled_data = data_wine.sample(frac=1,random_state=4)

# Поместите весь класс 1 в отдельный набор данных.
Class1_df = shuffled_data.loc[shuffled_data['quality'] == 'bad']

#Произвольно выберем 744 наблюдения из класса 0 (класс большинства)
Class0_df = shuffled_data.loc[shuffled_data['quality'] == 'good'].sample(n=744)

# Снова объедините оба фрейма данных
normalized_df = pd.concat([Class0_df, Class1_df])

# проверить количество новых классов
normalized_df.quality.value_counts()

print('Размер исходной выборки', data_wine.shape)
print('Размер выборки после андерсемплинга',normalized_df.shape)

#Проверим баланс выборки
sns.countplot(x='quality', data= normalized_df, palette="OrRd")
plt.box(False)
plt.xlabel('quality',fontsize=11)
plt.ylabel('Количество наблюдений',fontsize=11)
plt.show()
#Теперь выборка стала сбалансированной
```

```python
#Проведём оверсемплинг методом SMOTE (Способ Передискретизации Синтезированных Меньшинств)
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

print('Размер исходной выборки',X.shape)
print('Размер выборки после оверсемплинга',X_sm.shape)
df_new_sm=pd.concat([X_sm, y_sm], axis=1)


#Проверим баланс выборки
sns.countplot(x='quality', data= df_new_sm, palette="OrRd")
plt.box(False)
plt.xlabel('quality',fontsize=11)
plt.ylabel('Количество наблюдений',fontsize=11)
plt.show()
#В результате оверсемплинга выборка стала сбалансированной
```

```python
#Попробуем провести оверсемплинг с помощью RandomOverSampler
from imblearn.over_sampling import RandomOverSampler

sampling_strategy = 1
autopct = "%.2f"

ros = RandomOverSampler(sampling_strategy=sampling_strategy)
X_s, y_s = ros.fit_resample(X, y)

print('Размер исходной выборки',X.shape)
print('Размер выборки после оверсемплинга',X_s.shape)

ax = y_s.value_counts().plot.pie(autopct=autopct)
_ = ax.set_title("Over-sampling")
```

```python
#Сравним полученные результаты.
#В исходной выборке было выделено два класса:
#    -Класс0: Хорошие вина;
#    -Класс1: Плохие вина.
#Исходная выборка была изначально неплохо сбалансирована. Хороших и плохих вин было почти поровну.
#В рамках лабораторной работы по отношению к выборке были применены андерсемплинг и оверсемплинг.
#Оба подхода позволили получить сбалансированную выборку.
#Все использованные методы показали хорошие результаты и повысили сбалансированность выборки.
```
