```python
#Уберём из выборки записи, содержащие выбросы
data_wine_del = data_wine.copy()
data_wine_del = data_wine_del.drop(np.where((data_wine_del['fixed acidity'] < 4.6) | (data_wine['fixed acidity'] > 12.3))[0]) 	
data_wine_del = data_wine_del.drop(np.where((data_wine_del['volatile acidity'] < 0.12) | (data_wine['volatile acidity'] > 1.01))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['citric acid'] < 0) | (data_wine['citric acid'] > 0.79))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['residual sugar'] < 0.9) | (data_wine['residual sugar'] > 3.65))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['chlorides'] < 0.041) | (data_wine['chlorides'] > 0.12))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['free sulfur dioxide'] < 1) | (data_wine['free sulfur dioxide'] > 42))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['total sulfur dioxide'] < 6) | (data_wine['total sulfur dioxide'] > 122))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['density'] < 0.99235) | (data_wine['density'] > 1.001))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['pH'] < 2.93) | (data_wine['pH'] > 3.68))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['sulphates'] < 0.33) | (data_wine['sulphates'] > 1))[0])
data_wine_del = data_wine_del.drop(np.where((data_wine_del['alcohol'] < 8.4) | (data_wine['alcohol'] > 13.5))[0])
```

```python
#Проанализируем выборку на наличие выбросов
import plotly.graph_objs as go # графические объекты
fig = make_subplots(rows=4, cols=3)
# Add traces
fig.add_trace(go.Box(x=data_wine_del['fixed acidity'], boxpoints='all'), row=1, col=1)
fig.add_trace(go.Box(x=data_wine_del['volatile acidity'], boxpoints='all'), row=1, col=2)
fig.add_trace(go.Box(x=data_wine_del['citric acid'], boxpoints='all'), row=1, col=3)
fig.add_trace(go.Box(x=data_wine_del['residual sugar'], boxpoints='all'), row=2, col=1) 
fig.add_trace(go.Box(x=data_wine_del['chlorides'], boxpoints='all'), row=2, col=2)
fig.add_trace(go.Box(x=data_wine_del['free sulfur dioxide'], boxpoints='all'), row=2, col=3)
fig.add_trace(go.Box(x=data_wine_del['total sulfur dioxide'], boxpoints='all'), row=3, col=1)
fig.add_trace(go.Box(x=data_wine_del['density'], boxpoints='all'), row=3, col=2)
fig.add_trace(go.Box(x=data_wine_del['pH'], boxpoints='all'), row=3, col=3)
fig.add_trace(go.Box(x=data_wine_del['sulphates'], boxpoints='all'), row=4, col=1)
fig.add_trace(go.Box(x=data_wine_del['alcohol'], boxpoints='all'), row=4, col=2)
#fig.add_trace(go.Box(x=data_wine_del['quality'], boxpoints='all'), row=4, col=3)

# Update xaxis properties
fig.update_xaxes(title_text="fixed acidity", row=1, col=1)
fig.update_xaxes(title_text="volatile acidity", row=1, col=2)
fig.update_xaxes(title_text="citric acid", row=1, col=3)
fig.update_xaxes(title_text="residual sugar", row=2, col=1)
fig.update_xaxes(title_text="chlorides", row=2, col=2)
fig.update_xaxes(title_text="free sulfur dioxide", row=2, col=3)
fig.update_xaxes(title_text="total sulfur dioxide", row=3, col=1)
fig.update_xaxes(title_text="density", row=3, col=2)
fig.update_xaxes(title_text="pH", row=3, col=3)
fig.update_xaxes(title_text="sulphates", row=4, col=1)
fig.update_xaxes(title_text="alcohol", row=4, col=2)
#fig.update_xaxes(title_text="quality", row=4, col=3)

# Update yaxis properties
fig.update_yaxes(title_text="Amount", row=1, col=1)
fig.update_yaxes(title_text="Amount", row=1, col=2)
fig.update_yaxes(title_text="Amount", row=1, col=3)
fig.update_yaxes(title_text="Amount", row=2, col=1)
fig.update_yaxes(title_text="Amount", row=2, col=2)
fig.update_yaxes(title_text="Amount", row=2, col=3)
fig.update_yaxes(title_text="Amount", row=3, col=1)
fig.update_yaxes(title_text="Amount", row=3, col=2)
fig.update_yaxes(title_text="Amount", row=3, col=3)
fig.update_yaxes(title_text="Amount", row=4, col=1)
fig.update_yaxes(title_text="Amount", row=4, col=2)
#fig.update_yaxes(title_text="Amount", row=4, col=3)

fig.update_layout(title_text="Гистограммы по всем винным показателям", height=1000)
# Show the figure
fig.show()
#Выбросы удалены. Теперь оценим потери записей в выборке
print("Процент потерянных записей в выборке :",round((data_wine.shape[0] - data_wine_del.shape[0])/data_wine.shape[0]*100,2),"%.")
```
